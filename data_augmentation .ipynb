```python
import librosa 
import numpy as np 
import matplotlib.pyplot as plt 
import IPython.display as ipd 
import pandas as pd
from glob import glob
import os
from scipy.io.wavfile import write
```

```python
data_dir = os.listdir("./data/")
```

```python
j=0
ll = []

for d in data_dir:
    i = 0
    fs = glob("./data/"+d+"/*")
    for f in fs:
        i = i+1
  
        wav, sr = librosa.load(f,sr=22050)
        
        print("filesprocess: ", f , " -- " , i)
        wav_n = wav + 0.03*np.random.normal(0,1,len(wav))
        
        write("./data/"+d+ "/"+str(np.random.randint(0,99999999))+".m4a", sr , wav_n)
        print(" i est : {} \n".format(i))
        

        rock = -1.2
   
        wav_pitch_sf = librosa.effects.pitch_shift(wav,sr,n_steps=rock)
        write("./data/"+d+ "/"+str(np.random.randint(0,99999999))+"rock.m4a", sr , wav_pitch_sf)
        
        mince = 1.5
        wav_pitch_sf2 = librosa.effects.pitch_shift(wav,sr,n_steps=mince)
        write("./data/"+d+ "/"+str(np.random.randint(0,99999999))+"min.m4a", sr , wav_pitch_sf2)
    
```
